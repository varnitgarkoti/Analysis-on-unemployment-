```python
cleaned_df["Date"] = pd.to_datetime(cleaned_df["Date"], dayfirst = True)
```

```python
cleaned_df["Month"] = cleaned_df["Date"].dt.month
```

```python
cleaned_df["Frequency"] = cleaned_df["Frequency"].astype("category")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_corr = cleaned_df.corr(numeric_only = True)
#print(df_corr)
```

```python
## plt fig
plt.figure(figsize = (6,6))
plt.title("Correlation matix")
sns.heatmap(df_corr , annot = True)
plt.show()
```

```python
plt.title("Unemployment by region")
sns.barplot(x = "Region", y = "Estimated Unemployment Rate" , data = cleaned_df, palette= "winter")
plt.show()
```

```python
sns.scatterplot(
    x = "Estimated Employed",
    y = "Estimated Unemployment Rate",
    hue = "Region", 
    data = cleaned_df
)
```

```python
state_data = cleaned_df.groupby("Region")["Estimated Unemployment Rate"].mean().sort_values()
```

```python
plt.title("Distibution of Unemployment Rate")
plt.pie(
    state_data,
    labels = state_data.index,
    autopct = "%1.1f%%",
    startangle = 90,  
)

plt.show()
```

```python
plt.figure(figsize = (10 , 5))
sns.lineplot(
    x = "Date",
    y = "Estimated Unemployment Rate",
    hue = "Region",
    data = cleaned_df
)

plt.show()
```

analysis focuses on cleaning the unemployment dataset and performing EDA (correlation analysis, regional comparison, unemployment distribution, and trends over time). A concise 4-point conclusion could be:

Unemployment levels vary significantly across regions, indicating that employment opportunities are not evenly distributed throughout India.

The unemployment rate shows noticeable fluctuations over time, suggesting that economic conditions and external factors influence employment trends.

Some regions have higher labor participation and employment levels but still experience unemployment, highlighting the complexity of labor market dynamics.

The exploratory analysis provides valuable insights for policymakers, helping identify regions that may require targeted employment programs and economic support.
